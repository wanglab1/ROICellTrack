```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load the image
image_path = "example_A1_zoom.tiff"
image = Image.open(image_path)
# Convert image to numpy array
image_np = np.array(image)
```

```python
# Extract R, G, B channels and check color intensity
red_channel = image_np[:, :, 0]
green_channel = image_np[:, :, 1]
blue_channel = image_np[:, :, 2]

# Calculate mean intensity for each channel
mean_red_intensity = np.mean(red_channel)
mean_green_intensity = np.mean(green_channel)
mean_blue_intensity = np.mean(blue_channel)

# Create subplots to display the channels
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
# Display Red Channel
axes[0].imshow(red_channel, cmap='gray')
axes[0].set_title(f"Red (Mean color intensity: {mean_red_intensity:.2f})")
# Display Green Channel
axes[1].imshow(green_channel, cmap='gray')
axes[1].set_title(f"Green (Mean colorintensity: {mean_green_intensity:.2f})")
# Display Blue Channel
axes[2].imshow(blue_channel, cmap='gray')
axes[2].set_title(f"Blue (Mean color intensity: {mean_blue_intensity:.2f})")

plt.show()
```

```python
## Explore clustering analyiss for tissue compartment mining

from sklearn.cluster import KMeans

# Reshape the image for clustering
pixels = image_np.reshape(-1, image_np.shape[-1])  # Convert (H, W, C) to (N, C)

# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(pixels)
kmeans_labels = kmeans.labels_.reshape(image_np.shape[:2])  # Reshape to image dimensions

# Create subplots to display the original image and clustering results
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
# Original Image
axes[0].imshow(image_np)
axes[0].set_title("Original Image")
# K-Means Clustering
axes[1].imshow(kmeans_labels, cmap='jet')
axes[1].set_title("K-Means Clustering")

plt.show()
```
